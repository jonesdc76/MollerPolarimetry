```python
import csv
import matplotlib.pyplot as plt
import datetime as dt
import matplotlib.dates as md
from matplotlib import rc
rc('text', usetex=True)
plt.rcParams["figure.figsize"] = (15,7)
plt.rcParams.update({'font.size': 18})
with open("pita.csv") as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    dates_in = []
    dates_out = []
    pita_in = []
    pita_out = []
    pita_inV = []
    pita_outV = []
    for row in readCSV:
        hwp = row[2]
        if hwp == '1'and abs(float(row[3]))>1e-10:
            #print(row[1])
            pita_inV.append(float(row[3])*12670)
            pita_in.append(float(row[3])*3.1415926/2)
            dates_in.append(dt.datetime.fromtimestamp(float(row[1])))
        if hwp == '0' and abs(float(row[4]))>1e-10 and abs(float(row[4]))<0.1:
            pita_outV.append(float(row[4])*12670)
            pita_out.append(float(row[4])*3.1415926/2)
            dates_out.append(dt.datetime.fromtimestamp(float(row[1])))
    plt.xticks( rotation=70 )
    ax=plt.gca()
    xfmt = md.DateFormatter('%m-%d')
    ax.xaxis.set_major_formatter(xfmt)
    plt.plot(dates_in, pita_inV, ",")
    plt.title(r'IHWP In PITA Voltage versus Time: $\displaystyle\left(V_{PITA}-V{set}\right)$')
    plt.xlabel("Date in 2019")
    plt.ylabel("PITA Voltage (V)")
    plt.show()
    plt.xticks( rotation=70 )
    ax=plt.gca()
    ax.xaxis.set_major_formatter(xfmt)
    plt.title(r'IHWP Out PITA Voltage versus Time: $\displaystyle\left(V_{PITA}-V{set}\right)$')
    plt.xlabel("Date in 2019")
    plt.ylabel("PITA Voltage (V)")
    plt.plot(dates_out, pita_outV, ",",color='orange')
    plt.show()
```
